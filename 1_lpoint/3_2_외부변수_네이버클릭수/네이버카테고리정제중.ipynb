## naver 기준
- 패션의류 = 여성의류, 남성의류, 유아동의류, 속옷/'!양말'/홈웨어
- 패션잡화 = 패션잡화, 속옷/'양말'/홈웨어
- 스포츠/레저 = 구기/필드스포츠, 시즌스포츠, 스포츠패션, 헬스/피트니스, 아웃도어/레저
- 화장품/미용 = 화장품/뷰티케어, 퍼스널케어
- 디지털/가전 = 영상/음향가전, 컴퓨터, 생활/주방가전, 냉장/세탁가전, 계절가전, 모바일
- 가구/인테리어 = 가구, 인테리어/조명, 침구/수예
- 출산/육아 = 출산/육아용품 , 유아동의류, 완구
- 식품 = 과일, 음료, 냉동식품, 냉장식품, 축산물, 건강식품
- 생활/건강 = 청소/세탁/욕실용품, 주방잡화, 문구/사무용품, 세제/위생, 자동차용품, 원예/애완, 퍼스널케어(구강케어, 바디케어, 여성위생용품),식기/조리기구
- 여행/문화 = 상품권
