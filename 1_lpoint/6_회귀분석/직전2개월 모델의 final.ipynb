```python
from itertools import chain, combinations

def powerset(iterable):
    xs = list(iterable)
    return chain.from_iterable(combinations(xs,n) for n in range(len(xs)+1))

subsets = list(powerset(['F','M',10,20,30,40,50,60,'수도권','전라도','경상도','충청도','강원도','제주도']))
```

```python
moage = [60 if i >= 60 else i for i in subtotal['CLNT_AGE']]
subtotal['CLNT_AGE'] = moage
```

```python
zonname = []
for i in subtotal['ZON_NM']:
    if (i =='Seoul') | (i =='Gyeonggi-do') | (i=='Incheon'):
        zonname.append('수도권')
    elif (i == 'Jeollabuk-do') | (i =='Jeollanam-do') | (i=='Gwangju'):
        zonname.append('전라도')
    elif (i =='Busan') | (i =='Ulsan') | (i=='Gyeongsangnam-do')| (i =='Gyeongsangbuk-do') | (i=='Daegu'):
        zonname.append('경상도')
    elif (i =='Chungcheongnam-do') | (i =='Chungcheongbuk-do') | (i=='Daejeon'):
        zonname.append('충청도')
    elif i =='Gangwon-do':
        zonname.append('강원도')
    elif i =='Jeju-do':
        zonname.append('제주도')
    elif i == np.nan:
        zonname.append(np.nan)
        
```

```python
rmv = []
for sub in subsets:
    sltn = ('F','M')
    if all(item in sub for item in sltn):
        continue
    else: 
        rmv.append(sub)
            

rmv2 = []

for sub in rmv:
    sltn = ('수도권', '전라도', '경상도', '충청도', '강원도', '제주도')
    if all(item in sub for item in sltn):
        continue
    else: 
        rmv2.append(sub)

        
rmv3 = []

for sub in rmv2:   
    sltn = (10,20,30,40,50,60)
    if all(item in sub for item in sltn):
        continue
    else:
        rmv3.append(sub)
```

```python
M4 = M4.groupby(['CLNT_ID','SESS_ID','SESS_SEQ','TOT_SESS_HR_V','HITS_SEQ','CLAC3_NM','ZON_NM','CLNT_GENDER','CLNT_AGE','month','TOT_PAG_VIEW_CT']).sum()
```

```python
col_name = []
for sub in rmv3:
    val = [0 for i in range(len(sub))]
    for index in range(len(sub)):
        if sub[index] in ('F','M'):
            val[index] = 'CLNT_GENDER'
        elif sub[index] in (10,20,30,40,50,60):
            val[index] = 'CLNT_AGE'
        elif sub[index] in ('수도권','전라도','경상도','충청도','강원도','제주도'):
            val[index] = 'ZON_NM'
    col_name.append(val)        
    
```

```python
M4.reset_index(inplace=True)
```

```python
M4.head()
```

```python
final = []

for ind,key in enumerate(rmv3):
    print(ind)
    temp = M4
    
    gg = 0
    aa = 0
    zz = 0
    
    gq = ''
    aq = ''
    zq = ''
    
    for k,col in enumerate(col_name[ind]):
        if col == 'CLNT_GENDER':
            if gg >= 1:
                gq += '|' + 'CLNT_GENDER == "' + key[k]+ '"'
            else:    
                gq = 'CLNT_GENDER == "' + key[k]+ '"'
            gg += 1
            
        elif col == 'CLNT_AGE':
            if aa >= 1:
                aq += '|' + 'CLNT_AGE == "' + str(key[k])+'"'
            else:
                aq ='CLNT_AGE == "' + str(key[k])+'"'
            aa += 1
            
        elif col == 'ZON_NM':
            if zz >= 1:
                zq += '|' + 'ZON_NM == "' + str(key[k])+'"'
            else:
                zq = 'ZON_NM == "' + key[k]+ '"'
            zz += 1
    print(gq, aq, zq)
    
    if gg >= 1:
        temp = temp.query(gq)
    if aa >= 1:
        temp = temp.query(aq)
    if zz >= 1:
        temp = temp.query(zq)
    
    
    
    final.append(temp.groupby('CLAC3_NM').sum()[['PD_BUY_CT','HITS_SEQ']])
    
```

```python
final = []

for ind,key in enumerate(rmv3):
    print(ind)
    temp = M4
    
    gkey = []
    akey = []
    zkey = []
    for k,col in enumerate(col_name[ind]):
        if col == 'CLNT_GENDER':
            gkey.append(key[k])
        elif col == 'CLNT_AGE':
            akey.append(key[k])
        elif col =='ZON_NM':
            zkey.append(key[k])
            
    Gender = []
    Age = []
    Zon = []
    
    for gender in range(len(gkey)):
        #print(gkey)
        q = 'CLNT_GENDER == "' + gkey[gender] +'"'
        
        temp = temp.query(q)
        Gender.append(temp)
               
        if gender >= 1:
            print('안돼')
            temp = pd.concat([temp,Gender[gender]])
    
        
    for age in range(len(akey)):
        q = 'CLNT_AGE == "' + str(akey[age]) +'"'
        agetemp = temp.query(q)
        Age.append(agetemp)
        if age >= 1:
            temp2 = pd.concat([Age[age-1],Age[age]])
    
    if len(akey) >= 2: 
        temp = temp2
    elif len(akey) == 1:
        temp = agetemp
                
    for zon in range(len(zkey)):
        q = 'ZON_NM == "' + zkey[zon] +'"'
        zontemp = temp.query(q)
        Zon.append(zontemp)
        
        if zon >= 1:
            temp2 = pd.concat([Zon[zon-1],Zon[zon]])
    if len(zkey) >= 2: 
        temp = temp2     
    elif len(zkey) == 1:
        temp = zontemp
        
    final.append(temp)
    
                
```
