# 정규표현식으로 날씨 더 간단하게 만들기

```python
p = re.compile("^[가-힇]*")
result=[]
for i in weather.WEATHER:
    m = p.match(i)
    result.append(m.group())
```

```python
weather["simple"]=result
```

```python
weather.groupby("simple").count()
```

```python
sess_weath = session.merge(weather, left_on = ['SESS_DT','ZON_NM'], right_on=['DATE','ZON_NM'])
```

```python
sess_weath.groupby("simple").count()
```

```python
sess_weath.to_csv('session_weather.csv')
```
