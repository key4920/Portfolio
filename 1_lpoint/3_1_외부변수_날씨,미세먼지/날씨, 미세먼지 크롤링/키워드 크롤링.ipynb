```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
import csv
from urllib import parse
from itertools import count

def get_request_url(url,enc = 'utf-8'):
    req = urllib.request.Request(url)

    try:
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            try:
                rcv = response.read()
                ret = rcv.decode(enc)
            except UnicodeDecodeError:
                ret = rcv.decode(enc, 'replace')
            return ret
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
```

```python
data = pd.read_csv('podol_final2.csv')
```

```python
podolctgs = data[['CLAC3_NM','CLAC2_NM','CLAC1_NM']]
```

```python
podolctgs['CLAC1_NM'].unique()
```

```python
listt = podolctgs[(podolctgs['CLAC1_NM']=='패션잡화') | (podolctgs['CLAC1_NM']=='유아동의류') | (podolctgs['CLAC1_NM']=='여성의류') | (podolctgs['CLAC1_NM']=='아웃도어/레저') | (podolctgs['CLAC1_NM']=='시즌스포츠') | (podolctgs['CLAC1_NM']=='스포츠패션')]
```

```python
keywords = listt['CLAC3_NM']
```

```python
from urllib import parse

final = []
final_id = []
    
for keyword in keywords:
    print(keyword)
    if keyword[:2] =='기타':
        keyword = keyword[2:]
        print(keyword)
    
    url = parse.urlparse('https://search.shopping.naver.com/search/all.nhn?query=%s'% str(keyword) + '&cat_id=&frm=NVSHATC')

    query= parse.parse_qs(url.query)
    q = parse.urlencode(query, doseq = True)
    url = 'https://search.shopping.naver.com/search/all.nhn?' + q
    
    try:
        rcv_data = get_request_url(url)
        soupData = BeautifulSoup(rcv_data, 'html.parser')
        full_ctg = soupData.select('div.info > span.depth > a')
        ctgs = []
        ids = []
        for j in full_ctg:
            ctgs.append(j.text)
            ids.append(j.get('href'))
        for index,check in enumerate(ctgs):
            if check == keyword:
                print(check,'@@@@@')
                final.append(check)
                final_id.append(ids[index])
                break
            else:
                final.append(ctgs[-1])
                final_id.append(ids[-1][-8:])
    except:
        print('error with :%s'%keyword)
        final.append('nan')
        final_id.append('nan')
```
