```python
import pandas as pd
import datetime

pd.set_option('display.max_columns', None)
```

```python
dust = pd.read_csv('./날씨, 미세먼지 크롤링/dust.csv')
weather = pd.read_csv('./날씨, 미세먼지 크롤링/weather.csv')
```

```python
data = pd.read_csv('podol_subtotal.csv')
```

```python
dust['DATE'] = dust.DATE.apply(lambda x: datetime.datetime.strptime(str(x),'%Y%m%d'))
dpt = dust.DATE.dt
dpt_prop = pd.DataFrame({'year':dpt.year,
             'month':dpt.month,
             'day':dpt.day
             })
dust = pd.concat([dust,dpt_prop],axis=1)
dust.head()
```

```python
len(weather)
```

```python
dust[pd.isnull(dust['px_max'])]
```

```python
weather['DATE'] = weather.DATE.apply(lambda x: datetime.datetime.strptime(str(x),'%Y%m%d'))
wpt = weather.DATE.dt
wpt_prop = pd.DataFrame({'year':wpt.year,
             'month':wpt.month,
             'day':wpt.day
             })
weather= pd.concat([weather,wpt_prop],axis=1)
weather.head()
```

```python
data = data.merge(dust,on=['year','month','day','ZON_NM'],how='left')
data = data.merge(weather,on=['year','month','day','ZON_NM'],how='left')
```

```python
dust[dust['DATE']=='2018-05-24'] 
```

```python
data[pd.isnull(data['px_mean'])]
```

```python
data.drop(columns='DATE_x',inplace=True)
data.drop(columns='DATE_y',inplace=True)
```

```python
data.fillna(0,inplace=True)
```
