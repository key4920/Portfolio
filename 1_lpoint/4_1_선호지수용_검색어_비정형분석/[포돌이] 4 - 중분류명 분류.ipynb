# 중분류 기준 비율구하기

```python
data = pd.DataFrame(columns=['CLAC2_NM','CLAC3_NM', 'CLAC3_SALES'])

data.CLAC3_NM = pf.CLAC3_NM
data.CLAC2_NM = pf.CLAC2_NM
data.CLAC3_SALES = pf.TOT_SALES
data
```

```python
data2 = data.pivot_table(index=["CLAC2_NM", "CLAC3_NM"])
data2
```

```python
c2sum = data2.groupby(['CLAC2_NM']).sum()
```

```python
new = {}

for i in range(len(c2sum)):
    new[c2sum.index[i]] = c2sum['CLAC3_SALES'][i]
    
c2sum_dic = new
c2sum_dic
```

```python
list(c2sum_dic.keys())[0]
```

```python
data2['CLAC2_SALES']=0
```

```python
for j in range(len(c2sum_dic)):
    for i in range(len(data2)):
        if data2.index.get_level_values(0)[i] == list(c2sum_dic.keys())[j]:
            data2['CLAC2_SALES'][i] = list(c2sum_dic.values())[j]
    
data2
```

```python
data2['RATIO'] = 0
data2['RATIO'] = data2['RATIO'].astype(float)
```

```python
for i in range(len(data2)):
    data2['RATIO'][i] = (data2['CLAC3_SALES'][i] / data2['CLAC2_SALES'][i])
    
data2
```

# 중분류 Split

```python
long_data = data
```

```python
long_data['CP_CLAC2_NM'] = long_data['CLAC2_NM']
long_data.head()
```

```python
lendf = len(long_data)
for i in range(len(long_data)):
    if '/' in long_data['CLAC2_NM'][i]:
        temp = long_data['CLAC2_NM'][i].split('/')
        
        if '가구' in temp[len(temp)-1]:
            long_data['CLAC2_NM'][i] = temp[0]+'가구'
            
        elif '의류' in temp[len(temp)-1]:
            long_data['CLAC2_NM'][i] = temp[0]+'의류'
        
        elif '용기' in temp[len(temp)-1]:
            long_data['CLAC2_NM'][i] = temp[0]+'용기'
            
        else:
            long_data['CLAC2_NM'][i] = temp[0]

            
        for j in range(1,len(temp)):
            if '남성' in temp[0]:
                long_data.loc[lendf]= ['남성'+temp[j], long_data['CLAC3_NM'][i], 
                                       long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
            elif '주방' in temp[0]:
                long_data.loc[lendf]= ['주방'+temp[j], long_data['CLAC3_NM'][i], 
                                       long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
            elif '자동차' in temp[0]:
                long_data.loc[lendf]= ['자동차'+temp[j], long_data['CLAC3_NM'][i], 
                                       long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
            
            else:
                long_data.loc[lendf]= [temp[j], long_data['CLAC3_NM'][i], 
                                       long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
                         
            lendf += 1
      
```

```python
lendf = len(long_data)
for i in range(len(long_data)):
    strtemp = ''
    
    if '일반' in long_data['CLAC2_NM'][i] and '의류' in long_data['CLAC2_NM'][i]:        
        temp = long_data['CLAC2_NM'][i].split('일반')    
        for k in range(len(temp)):
            if '일반' not in temp[k]:
                strtemp =  strtemp + temp[k]
            else:
                temp[k] = ''
                strtemp =  strtemp + temp[k]
                
        temp2 = strtemp.split('의류')
        strtemp2 = ''
        for k in range(len(temp2)):
            if '의류' not in temp2[k]:
                strtemp2 =  strtemp2 + temp2[k]
            else:
                temp2[k] = ''
                strtemp2 =  strtemp2 + temp2[k]
         
        
        long_data.loc[lendf]= [strtemp2, long_data['CLAC3_NM'][i], 
                               long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
        lendf += 1
            
    elif '일반' in long_data['CLAC2_NM'][i]:
        temp = long_data['CLAC2_NM'][i].split('일반')
        for k in range(len(temp)):
            if '일반' not in temp[k]:
                strtemp =  strtemp + temp[k]
            else:
                temp[k] = ''
                strtemp =  strtemp + temp[k]
        
        long_data.loc[lendf]= [strtemp, long_data['CLAC3_NM'][i], 
                               long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
        lendf += 1
    
    elif '의류' in long_data['CLAC2_NM'][i] and long_data['CLAC2_NM'][i] != '아웃도어의류':
        temp = long_data['CLAC2_NM'][i].split('의류')
        for k in range(len(temp)):
            if '의류' not in temp[k]:
                strtemp =  strtemp + temp[k]
            else:
                temp[k] = ''
                strtemp =  strtemp + temp[k]
                
        #print(long_data['CLAC2_NM'][i],strtemp)      
        long_data.loc[lendf]= [strtemp, long_data['CLAC3_NM'][i], 
                               long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
        lendf += 1
    
    elif '양산류' in long_data['CLAC2_NM'][i] or '양말류' in long_data['CLAC2_NM'][i] or '가방류' in long_data['CLAC2_NM'][i]:
        temp = long_data['CLAC2_NM'][i].split('류')
        for k in range(len(temp)):
            if '류' not in temp[k]:
                strtemp =  strtemp + temp[k]
            else:
                temp[k] = ''
                strtemp =  strtemp + temp[k]
                
        #print(long_data['CLAC2_NM'][i],strtemp)      
        long_data.loc[lendf]= [strtemp, long_data['CLAC3_NM'][i], 
                               long_data['CLAC3_SALES'][i], long_data['CP_CLAC2_NM'][i]]
        lendf += 1
```

```python
detail = {}
for i in range(len(detail_long_data2_dict)):
    key = list(detail_long_data2_dict.keys())[i]
    value = list(detail_long_data2_dict.values())[i]
    strtemp = ''
    if '남성' in key and key != '남성화':
        temp = key.split('남성')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['남성 ' + strtemp] = value; detail[key] = value
    
    elif '냉동' in key and '냉동고' != key:
        temp = key.split('냉동')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['냉동 ' + strtemp] = value; detail[key] = value
    
    elif '주방' in key:
        temp = key.split('주방')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['주방 ' + strtemp] = value; detail[key] = value
        
        
    elif '건강' in key:
        temp = key.split('건강')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['건강 ' + strtemp] = value; detail[key] = value
        
    elif '컴퓨터' in key:
        temp = key.split('컴퓨터')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['컴퓨터 ' + strtemp] = value; detail[key] = value
    
    elif '여성' in key and key != '여성화':
        temp = key.split('여성')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['여성 ' + strtemp] = value; detail[key] = value
        
    elif '남아' in key:
        temp = key.split('남아')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['남아 ' + strtemp] = value; detail[key] = value

    elif '남아' in key:
        temp = key.split('남아')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['남아 ' + strtemp] = value; detail[key] = value

    elif '여아' in key:
        temp = key.split('여아')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['여아 ' + strtemp] = value; detail[key] = value
     
    elif '유아동' in key and key != '유아동화':
        temp = key.split('유아동')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['유아동 ' + strtemp] = value; detail[key] = value
    
    elif '유아' in key:
        temp = key.split('유아')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['유아 ' + strtemp] = value; detail[key] = value
    
    elif '자동차' in key:
        temp = key.split('자동차')
        for i in range(1,len(temp)):
            strtemp = strtemp + temp[i]
        detail['자동차 ' + strtemp] = value; detail[key] = value
      
    else:
        detail[key] = value
    
detail    
```
