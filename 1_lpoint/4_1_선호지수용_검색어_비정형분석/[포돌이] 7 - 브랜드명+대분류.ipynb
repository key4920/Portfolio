```python
pf = pd.read_csv('./csv/podol_final2.csv').drop('Unnamed: 0', axis= 1)
```

```python
bigdata = pd.DataFrame(columns=['CLAC1_NM','CLAC3_NM', 'CLAC3_SALES'])

bigdata.CLAC3_NM = pf.CLAC3_NM
bigdata.CLAC1_NM = pf.CLAC1_NM
bigdata.CLAC3_SALES = pf.TOT_SALES
bigdata.head()
```

```python
bigdata2 = bigdata.pivot_table(index=["CLAC1_NM", "CLAC3_NM"])
bigdata2
```

```python
c1sum = bigdata2.groupby(['CLAC1_NM']).sum()
```

```python
new = {}

for i in range(len(c1sum)):
    new[c1sum.index[i]] = c1sum['CLAC3_SALES'][i]
    
c1sum_dic = new
c1sum_dic
```

```python
bigdata2['CLAC1_SALES']=0
for j in range(len(c1sum_dic)):
    for i in range(len(bigdata2)):
        if bigdata2.index.get_level_values(0)[i] == list(c1sum_dic.keys())[j]:
            bigdata2['CLAC1_SALES'][i] = list(c1sum_dic.values())[j]
    
bigdata2
```

```python
bigdata2['RATIO'] = 0
bigdata2['RATIO'] = bigdata2['RATIO'].astype(float)
```

```python
for i in range(len(bigdata2)):
    bigdata2['RATIO'][i] = (bigdata2['CLAC3_SALES'][i] / bigdata2['CLAC1_SALES'][i])
    
bigdata2
```

```python
bigdata2.to_csv('CLAC1_NM_RATIO.csv')
bigdata2.to_csv('./csv/CLAC1_NM_RATIO.csv', encoding='CP949')
```

# 대분류 Split

```python
long_data = bigdata
```

```python
long_data['CP_CLAC1_NM'] = long_data['CLAC1_NM']
long_data.head()
```

```python
lendf = len(long_data)
for i in range(len(long_data)):
    if '/' in long_data['CLAC1_NM'][i]:
        temp = long_data['CLAC1_NM'][i].split('/')
        
        if '가전' in temp[len(temp)-1]:
            long_data['CLAC1_NM'][i] = temp[0]+'가전'
            
        elif '스포츠' in temp[len(temp)-1]:
            long_data['CLAC1_NM'][i] = temp[0]+'스포츠'
        
        elif '용품' in temp[len(temp)-1]:
            long_data['CLAC1_NM'][i] = temp[0]+'용품'
            
        else:
            long_data['CLAC1_NM'][i] = temp[0]

            
        for j in range(1,len(temp)):
            long_data.loc[lendf]= [temp[j], long_data['CLAC3_NM'][i], 
                                       long_data['CLAC3_SALES'][i], long_data['CP_CLAC1_NM'][i]]
                         
            lendf += 1
```

```python
long_data
```

```python
long_c1sum = long_data2.groupby(['CP_CLAC1_NM']).sum()
long_c1sum.head()
```

```python
long_new = {}

for i in range(len(long_c1sum)):
    long_new[long_c1sum.index[i]] = long_c1sum['CLAC3_SALES'][i]
    
long_c1sum_dic = long_new
long_c1sum_dic
```

```python
long_data2['RATIO'] = 0
long_data2['RATIO'] = long_data2['RATIO'].astype(float)
```
