# 분류된 데이터 불러오기

```python
import pandas as pd

data0 = pd.read_csv('./zerotoeight/index_data_0_to_8.csv')
data8 = pd.read_csv('./eighttofifteen/index_data_8_to_15.csv')
data15 = pd.read_csv('./fifteento23/index_data_15_to_23.csv')
data23 = pd.read_csv('./threetosix/index_data_23_to_26.csv')
data23.head()
```

```python
print(len(data0),len(data8),len(data15),len(data23))
```

```python
data0.head()
```

# 분류된 데이터 search_complete2으로 저장

```python
temp1 = pd.concat([data0, data8])
temp2 = pd.concat([temp1, data15])
search_complete2 = pd.concat([temp2, data23])
search_complete2
```

```python
search_complete2[search_complete2['CLAC3_NM'] == '0']
```

```python
search_complete2.to_csv('./result/seach_complete2.csv')
search_complete2.to_csv('./result/seach_complete2_cp949.csv', encoding='CP949')
```
