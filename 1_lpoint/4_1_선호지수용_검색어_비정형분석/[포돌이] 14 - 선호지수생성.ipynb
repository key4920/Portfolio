```python
import pandas as pd
import numpy as np

pf = pd.read_csv('./csv/podol_final2.csv').drop('Unnamed: 0', axis= 1)
search = pd.read_csv('clac3_sum_search2.csv').drop('Unnamed: 0', axis= 1)
```

```python
search
```

```python
data = pd.DataFrame()
```

```python
data['CLAC3_NM'] = pf.CLAC3_NM
data['TOT_SALES'] = pf.TOT_SALES
data['AVG_HITS'] = pf.AVG_HITS
```

```python
data
```

```python
mg = pd.merge(data, search, on='CLAC3_NM')
mg
```

```python
mg.describe()
```

```python
import numpy as pd
criteria = [mg['TOT_SALES'].between(2.000000,1781.000000), mg['TOT_SALES'].between(1781.000000, 5533.000000), mg['TOT_SALES'].between(5533.000000, 330277)]
values = ['L', 'M', 'H']

mg['TOT_SALES_LEVEL'] = np.select(criteria, values, 0)
```

```python
criteria2 = [mg['AVG_HITS'].between(16.750000, 68.035000), mg['AVG_HITS'].between(68.035000, 79.512500), mg['AVG_HITS'].between(79.512500, 175)]
values2 = ['L', 'M', 'H']

mg['AVG_HITS_LEVEL'] = np.select(criteria2, values2, 0)
```

```python
criteria3 = [mg['ratio'].between(0, 681.699953), mg['ratio'].between(681.699953, 2989.421622), mg['ratio'].between(2989.421622, 212791)]
values3 = ['L', 'M', 'H']

mg['TOT_SEARCH_LEVEL'] = np.select(criteria3, values3, 0)
```

```python
from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=['L', 'M', 'H'], ordered=True)
```

```python
mg['TOT_SALES_LEVEL'] = mg['TOT_SALES_LEVEL'].astype(cat_type)
mg['TOT_SEARCH_LEVEL'] = mg['TOT_SEARCH_LEVEL'].astype(cat_type)
mg['AVG_HITS_LEVEL'] = mg['AVG_HITS_LEVEL'].astype(cat_type)
```

```python
mg.sort_values(['TOT_SALES_LEVEL', 'TOT_SEARCH_LEVEL','AVG_HITS_LEVEL'], ascending=False, inplace=True)
```

```python
mg
```

```python
mg.to_csv('sort_preference.csv')
mg.to_csv('sort_preference_cp949.csv',encoding='CP949')
```

## 선호지수 생성

```python
import pandas as pd
bc = pd.read_csv('./csv/btwness_centrality.csv') #네트워크 데이터
bc.head()
```

```python
bc = bc.rename(columns = {'Unnamed: 0':'CLAC3_NM'})
bc.head()
```

```python
mg.head()
```

```python
mg = mg.rename(columns = {'ratio':'TOT_SEARCH_CNT'})
```

```python
mg2 = pd.merge(mg, bc, on='CLAC3_NM')
mg2
```

```python
from sklearn.preprocessing import scale
mg2['SCALE_TOT_SALES'] = scale(mg2['TOT_SALES'])
mg2['SCALE_AVG_HITS'] = scale(mg2['AVG_HITS'])
mg2['SCALE_TOT_SEARCH_CNT'] = scale(mg2['TOT_SEARCH_CNT'])
mg2['SCALE_BC'] = scale(mg2['BC'])
mg2
```

```python
mg2['PREFRENCE_IDX'] = (mg2['SCALE_TOT_SALES'] * 5) + (mg2['SCALE_TOT_SEARCH_CNT'] * 3) + (mg2['SCALE_BC'] * 2)
```

```python
mg3 = mg2.sort_values(by='PREFRENCE_IDX', ascending=False)
mg3
```

```python
mg4 = mg3.reindex_axis(['CLAC3_NM','PREFRENCE_IDX','TOT_SALES_LEVEL','TOT_SEARCH_LEVEL','AVG_HITS_LEVEL','TOT_SALES','AVG_HITS','TOT_SEARCH_CNT','BC','SCALE_TOT_SALES','SCALE_TOT_SEARCH_CNT','SCALE_AVG_HITS','SCALE_BC'],axis=1)
mg4
```

```python
mg4 = mg4.reset_index().drop('index', axis=1)
mg4
```

```python
mg4['TOT_SALES_LEVEL'] = mg4['TOT_SALES_LEVEL'].astype(cat_type)
mg4['TOT_SEARCH_LEVEL'] = mg4['TOT_SEARCH_LEVEL'].astype(cat_type)
mg4['AVG_HITS_LEVEL'] = mg4['AVG_HITS_LEVEL'].astype(cat_type)
```

```python
from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=['L', 'M', 'H'], ordered=True)
```

```python
mg4.sort_values(['TOT_SALES_LEVEL', 'TOT_SEARCH_LEVEL','AVG_HITS_LEVEL'], ascending=False, inplace=True)
mg4
```

=> 위에서부터 차례대로 선호지수가 높다
