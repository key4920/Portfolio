# 박광렬조에게 네트워크 대충 설명ㅋㅋ
 - 1거래의 정의 = '[CLNT_ID + SESS_ID + HITS_SEQ]' 고객이 세션에서 구매당시 hit가 
 - row 한 쌍이 같이 거래된 상품 (node 1, node2) --> 네트워크에서 edge가 됨 
 - 같은 쌍이 많을 수록 edge가 강해지는 것 
 - 현재 순서는 정렬 안됨 '네트워크 분석' 파일에 순서 정렬해서 그래프 생성해둠
