```python
dir = "./data_version2/"
datasets = {}

datasets['geolocation'] = pd.read_csv(dir + 'geolocation_olist_public_dataset.csv', dtype = {'zip_code_prefix' : str})
datasets['classified'] = pd.read_csv(dir + 'olist_classified_public_dataset.csv', dtype = {'customer_zip_code_prefix' : str},
                         parse_dates = ['order_purchase_timestamp', 'order_aproved_at', 'order_estimated_delivery_date', 'order_delivered_customer_date', 'review_creation_date', 'review_answer_timestamp'])
datasets['dataset'] = pd.read_csv(dir + 'olist_public_dataset_v2.csv', dtype = {'customer_zip_code_prefix' : str},
                      parse_dates = ['order_purchase_timestamp', 'order_aproved_at', 'order_estimated_delivery_date', 'order_delivered_customer_date', 'review_creation_date', 'review_answer_timestamp'])
datasets['customers'] = pd.read_csv(dir + 'olist_public_dataset_v2_customers.csv')
datasets['payments'] = pd.read_csv(dir + 'payments_olist_public_dataset.csv')
datasets['translation'] = pd.read_csv(dir + 'product_category_name_translation.csv')
datasets['product_measures'] = pd.read_csv(dir + 'product_measures_olist_public_dataset_.csv')
datasets['sellers'] = pd.read_csv(dir + 'sellers_olist_public_dataset_.csv', dtype = {'seller_zip_code_prefix' : str})
```
